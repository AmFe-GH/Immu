```python
import glob
import os
folder_path = '../Data/result_best_performance/S_level/'

# 获取文件夹下所有CSV文件
csv_files = glob.glob(os.path.join(folder_path, '*_NGS_ranking.csv'))
S_list = []
# 遍历每个CSV文件
for file in csv_files:
    # 读取CSV文件
    df = pd.read_csv(file, index_col=0)
    
    # 处理数据：取第十行及其以后的数据，填充NaN为0
    df_processed = df.iloc[10:].fillna(0)
    
    # 生成权重系数数组
    weights = (50 / 40) ** np.arange(len(df_processed))
    
    # 按行乘以权重系数
    df_weighted = df_processed.multiply(weights, axis=0)
    
    # 累加计算并取最后一行排序
    result = df_weighted.cumsum().iloc[-1].sort_values(ascending=False)
    S_list.append(result)
```

```python

import pandas as pd

# 假设 T_list 是一个包含多个 DataFrame 或 Series 的列表
# 首先将 T_list 中的所有元素相加
multi_result = sum(S_list)

# 确保 multi_result 是一个 DataFrame，如果它是 Series，转换成 DataFrame
if isinstance(multi_result, pd.Series):
    multi_result = multi_result.to_frame('importance_score')

# 将 multi_result 排序
multi_result = multi_result.sort_values(by='importance_score', ascending=False)

# 确保 DataFrame 有两列，'feature_name' 和 'importance_score'
multi_result.reset_index(inplace=True)
multi_result.columns = ['feature_name', 'importance_score']

# 保存为 CSV 文件
multi_result.to_csv('../Data/result_best_performance/S_level/NGS_marker.csv', index=False)

# 显示前20行
print(multi_result.head(20))
```

# G level

```python
import glob
import os
import pandas as pd
import numpy as np
level_list = ['T','S','G','F']
for level in level_list:
    folder_path = f'../Data/result_best_performance/{level}_level/'
    os.makedirs(folder_path,exist_ok=True)
    # 获取文件夹下所有CSV文件
    csv_files = glob.glob(os.path.join(folder_path, 'trial*_NGS_ranking.csv'))
    G_list = []
    # 遍历每个CSV文件
    for file in csv_files:
        # 读取CSV文件
        df = pd.read_csv(file, index_col=0)
        
        # 处理数据：去除第一行，取第十行及其以后的数据，填充NaN为0
        df_processed = df.drop(index=0).iloc[10:].fillna(0)
        
        # 生成权重系数数组
        weights = (1 / 0.8) ** np.arange(1, len(df_processed) + 1)
        
        # 按行乘以权重系数
        df_weighted = df_processed.multiply(weights, axis=0)
        
        # 累加计算并取最后一行排序
        result = df_weighted.cumsum().iloc[-1].sort_values(ascending=False)
        G_list.append(result)

    # 假设 T_list 是一个包含多个 DataFrame 或 Series 的列表
    # 首先将 T_list 中的所有元素相加
    multi_result = sum(G_list)

    # 确保 multi_result 是一个 DataFrame，如果它是 Series，转换成 DataFrame
    if isinstance(multi_result, pd.Series):
        multi_result = multi_result.to_frame('importance_score')

    # 将 multi_result 排序
    multi_result = multi_result.sort_values(by='importance_score', ascending=False)

    # 确保 DataFrame 有两列，'feature_name' 和 'importance_score'
    multi_result.reset_index(inplace=True)
    multi_result.columns = ['feature_name', 'importance_score']

    # 保存为 CSV 文件
    multi_result.to_csv(f'../Data/result_best_performance/{level}_marker_new.csv', index=False)

    # 显示前20行
    print(multi_result.head(20))
```

```python
import glob
import os
import pandas as pd
import numpy as np
level_list = ['S']
for level in level_list:
    folder_path = f'../Data/result_best_performance/meta/'
    os.makedirs(folder_path,exist_ok=True)
    # 获取文件夹下所有CSV文件
    csv_files = glob.glob(os.path.join(folder_path, 'trial*_meta_ranking.csv'))
    assert len(csv_files)!=0,"This folder does not contain any files" \
        ", please manually copy the trial*_meta_ranking.csv to this folder"
    
    G_list = []
    # 遍历每个CSV文件
    for file in csv_files:
        if 'trial24' in file or 'trial56' in file:
            reduction = 18
        else:
            reduction = 19
        # 读取CSV文件
        df = pd.read_csv(file, index_col=0)
        
        G_list.append(df.iloc[reduction].dropna())

df_list = []
for index,item in enumerate(G_list):
    item = item.reset_index()
    item.columns = ['features','Scores']
    G_list[index] = item
concat = G_list[0]
for s in G_list[1:]:
    concat = pd.concat([concat,s])

concat = concat.groupby('features')['Scores'].sum().reset_index()
concat.to_csv('../Data/result_best_performance/meta/meta_markers.csv')
# 输出结果
print(concat)
# # 转换为 Series 并输出结果
# print(result)

# G_list
```
